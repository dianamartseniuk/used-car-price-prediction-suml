# Data exploration

Basic exploration of the Poland cars for sale dataset.

```python
from pathlib import Path
import sys

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

from src.data_loading import load_raw_data

pd.set_option('display.max_columns', 100)
sns.set_theme(style='whitegrid')
```

```python
df = load_raw_data()
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
pd.DataFrame({
    'missing_count': df.isna().sum(),
    'missing_percent': (df.isna().mean() * 100).round(2),
    'dtype': df.dtypes.astype(str)
}).sort_values('missing_percent', ascending=False)
```

```python
print('Duplicated rows:', df.duplicated().sum())
print('Target column:', 'Price')
df['Price'].describe()
```

```python
important_categories = ['Vehicle_brand', 'Vehicle_model', 'Fuel_type', 'Transmission', 'Type', 'Drive', 'Condition', 'Origin_country']
for column in important_categories:
    if column in df.columns:
        print(f'\n{column}: {df[column].nunique(dropna=True)} unique values')
        display(df[column].value_counts(dropna=False).head(15))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 4))
sns.histplot(df['Price'].dropna(), bins=80, ax=axes[0])
axes[0].set_title('Price distribution')
axes[0].set_xlim(0, df['Price'].quantile(0.99))

sns.histplot(df['Mileage_km'].dropna(), bins=80, ax=axes[1])
axes[1].set_title('Mileage distribution')
axes[1].set_xlim(0, df['Mileage_km'].quantile(0.99))

sns.histplot(df['Production_year'].dropna(), bins=40, ax=axes[2])
axes[2].set_title('Production year distribution')
plt.tight_layout()
```

```python
numeric_columns = ['Price', 'Mileage_km', 'Production_year', 'Power_HP', 'Displacement_cm3']
df[numeric_columns].describe(percentiles=[0.01, 0.05, 0.5, 0.95, 0.99]).T
```

## Notes for cleaning and features

Useful features for the first model: brand, model, production year, mileage, power, engine capacity, fuel type, transmission, body type, drive, condition, doors, colour, origin country, first owner, and location.

Columns like index, publication date, exact registration date, very detailed features text, and exact location text should be simplified or removed for the first version.

The target `Price` is skewed, so training on `log1p(Price)` should make the regression problem easier.
